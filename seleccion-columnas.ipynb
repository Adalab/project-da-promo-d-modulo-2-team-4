```python
df_data_kaggle = pd.read_csv("archivos_originales/_data_remaning_kaggle.csv")
```

```python
df_data_kaggle.head(2)
```

```python
def seleccionar_columnas(df, patron):
    columnas = df.columns
    lista_columnas = ["index"]
    for col in columnas:
        elemento_col = re.findall(patron, col)
        if len(elemento_col) ==1:
            lista_columnas.append(elemento_col[0])
    df = df[lista_columnas]
    return df
  
```

```python
df_data_kaggle = seleccionar_columnas(df_data_kaggle, "Q39.*|Q32_B.*|Q34_B.*")
```

```python
df_data_kaggle.head(2)
```

```python
df_xml = pd.read_xml("archivos_originales/_data_xml_clean(1).xml")
```

```python
df_xml.head()
```

```python
df_xml = seleccionar_columnas(df_xml, "age|gender")
```

```python
df_txt = seleccionar_columnas(df_txt, "Q3[^2]|Q3$|Q41|Q4|Q5|Q6|Q7|Q8|Q9|Q1[^1|^3]|Q20|Q2[^2|^6]")
```

```python
df_total = df_txt.merge(df_xml, on ="index")
df_total = df_total.merge(df_data_kaggle, on="index")
df_total.head(1)
```
