```python
def seleccionar_columnas(df, patron):
    columnas = list(df.columns)
    lista_columnas = ["index"]
    for col in columnas:
        elemento_col = re.findall(patron, col)
        if len(elemento_col) ==1:
            lista_columnas.append(elemento_col[0])
    df = df[lista_columnas]
    return df
  
```

```python
df_data_kaggle = seleccionar_columnas(df_data_kaggle, "Q39.*|Q32_B.*|Q34_B.*")
```
