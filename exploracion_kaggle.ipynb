```python
columnas = df_data_kaggle.columns
columnas = list(columnas)
```

```python
lista_columnas = ["index"]
patron1 = "Q39.*|Q32_B.*|Q34_B.*"

for col in columnas:
    col_39 = re.findall(patron1, col)
    if len(col_39) ==1:
        lista_columnas.append(col_39[0])
    #print(col_39)
    #print(".........")
```

```python
df_kaggle_edit = df_data_kaggle[lista_columnas]
```
